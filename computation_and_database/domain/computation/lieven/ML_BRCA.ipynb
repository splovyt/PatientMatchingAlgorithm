```python
import pandas
import numpy as np
import sklearn
```

```python
data = pandas.read_table('BRCA_for_machine_learning.txt', sep='\t')
print(data.shape)
data
```

```python
# split test and train
from sklearn.model_selection import train_test_split
train, test = train_test_split(data, test_size = 0.4)

# seperate observations (Y) from features (X)
Y_train = train['subtype'] # observations
X_train = train.drop('subtype', axis=1) # features

Y_test = test['subtype'] # observations
X_test = test.drop('subtype', axis=1) # features
```

```python
# import the class
from sklearn.linear_model import LogisticRegression

# instantiate the model (using the default parameters)
logreg = LogisticRegression()

# fit the model with data
logreg.fit(X_train, Y_train)

target_Y = pandas.DataFrame(logreg.predict(X_test))[0].values.tolist()
target_Y = list(map(int, target_Y))
true_Y = list(Y_test)
true_Y = list(map(int, true_Y))
size = len(true_Y)
correct = 0
for i in range(size):
    if target_Y[i] == true_Y[i]:
        correct += 1
print(correct/size)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(X_train, Y_train)
print('training accuracy: ', clf.score(X_train, Y_train))
print('testing accuracy: ', clf.score(X_test, Y_test))

import sklearn.metrics
from sklearn.model_selection import cross_val_score
target_Y = pandas.DataFrame(clf.predict(X_test))[0].values.tolist()
target_Y = list(map(int, target_Y))
true_Y = list(Y_test)
true_Y = list(map(int, true_Y))
print('CrossValidation: ', np.mean(cross_val_score(clf, X_test, Y_test, cv=10)))
```

```python
from sklearn.svm import SVC
clf = SVC()
clf.fit(X_train, Y_train)
print('training accuracy: ', clf.score(X_train, Y_train))
print('testing accuracy: ', clf.score(X_test, Y_test))

import sklearn.metrics
from sklearn.model_selection import cross_val_score
target_Y = pandas.DataFrame(clf.predict(X_test))[0].values.tolist()
target_Y = list(map(int, target_Y))
true_Y = list(Y_test)
true_Y = list(map(int, true_Y))
print('CrossValidation: ', np.mean(cross_val_score(clf, X_test, Y_test, cv=10)))
```
